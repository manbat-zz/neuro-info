# Meta-Analysis For NeuroImaging

In neuroimaging analysis, there are multiple levels of "inference."  This basically just means modeling of your data (magnetic resonance imaging runs, we can focus on functional to limit the scope) so that I can "say something intelligent" (aka infer) about the larger population. The strategies that we choose on each level for doing these models must be statistically sound in a way so that we can make inferences that we want.  So, here is a very basic decomposition of the levels of a simple analysis with choices in the context of the inferences that we want to make:

## Levels of an Analysis
 - Single subjects: we typically might combine lots of runs from the same person to average out noise, and things like that.  We use [fixed-effects](http://en.wikipedia.org/wiki/Fixed_effects_model) models because variance within that person isn't interesting. 
 - Group analysis: we moved to [random-effects](http://en.wikipedia.org/wiki/Random_effects_model), because we want to make inferences about the population at large.  For the single subject analysis, our "units" that went in were the different runs. We assumed variance between those runs was not interesting.  For this next level, our "units" are the entire people.  Variance between people is interesting. We include it in the model because it is those differences we want to make inferences about.
 
Guess what - the procedure of using fixed-effects on the first level, and random-effects on the second has a name! We call it *drumroll* [mixed-effects](http://en.wikipedia.org/wiki/Mixed_model). So if you are doing your analysis in SPM, or FSL, or AFNI, and (let's hope you aren't using Brain Voyager), you have preprocessed and analyzed your single subjects, done a group analysis, and *poof* out comes a statistical map!  What do you do next? You threshold, correct for multiple comparisons, find some significant voxels, and write it up in a paper.

### But hold the phone, has anyone else found this?
You probably don't just write the paper mindlessly. You open up your map in FSLview, load some anatomical atlases, and look at the region names.  Then you do a Google search to see if anyone else has found that.  You spend lots of time and get tired, and say, "Wow, this is tiring... is there a better way?"

## Coordinate Based vs Image Based Meta Analysis
The current neuroimaging results-reporting landscape is about the peaks.  Everyone writes about the maximum voxels of activation in there papers, they put them in tables, and show how pretty they are in reds and oranges in brain figures. This has led to the development of methods for "meta analysis" that are based on coordinate data, "coordinate-based meta analysis" (CBMA). Some [players in the game](http://vsoch.com/blog/2015/01/visual-explanation-of-coordinate-based-meta-analysis-approaches/) are Kernel Density Analysis (KDA), Activation Likelihood Estimate (ALE), and Multilevel Kernel Density Analysis (MKDA). What these methods can do is take a whole bunch of reported brain coordinates, and look for significant overlap / clustering in the spatial patterns by comparing to a null distribution of randomly generated points in that same space. This is great given that we only have these peaks, but the landscape is changing! We are [starting to share more whole brain maps](http://www.neurovault.org), and this is exciting, because now we have mountains and not just peaks. So let's step back for a second.  This meta-analysis is really just one more step up on this ladder of inference.  Now the "units" are entire studies, and we want to make inferences about our understanding of the brain and cognition based on these maps. This second kind of meta-analysis is called Image Based Meta Analysis (IBMA), and you can imagine the simplest route would be to take some average over the group maps.  Let's put this third level of inference in the context of the hierarchy we defined previously:

|   |Units|Strategy|Questions|
|---|---|---|---|
|Level 1|runs for one guy|fixed-effects|where does this person's brain respond to cats?|
|Level 2|single subjects|random-effects|where does group X's brain respond to cats?|
|Level 3|an entire group|?|what do we know about the human experience of seeing a cat?|

and don't forget that fixed-effects + random-effects = "hierarchical mixed-effects model"

## Goals of the  Inference
The example above may be a little silly, but it hints at the goal of inference: to choose methods that let us answer the questions we want to ask! Here are some examples of questions that might be addressed by this third level of inference (going from simple to more complicated). Also remember that the researcher may or may not have information about the map of interest he/she wants to make inferences about, so our new method should be flexible to that:
 - Has anyone else found this result?
 - What is the probability that my region(s) (separately) are activated for different cognitive processes?
 - What is the probability that the (entire map) is indicative of a cognitive processs?
 - What is the likelihood of this result, based on previous results. Given my task?
 - If I see activation in region(s) X, does that mean cognitive process Z was going on?
 - If I know my task is a particular *kind* of emotion, can I understand how active the amygdala is in different kinds of emotions? The "Constructionist View" says that basic categories of emotion arise from complex combinations of elemental information processing operations across the brain, so it would be valuable to be able to make a more detailed inference beyond the category of "emotion" (e.g., if we have contrasts for an emotion study, our method should be able to take advantage of that to provide an even more detailed inference).

What we really are getting at here is "How does my result fit in the context of the overwhelming compendium of previous literature?" This is important to know because the entire purpose of non-invasive brain imaging is to understand how the brain works.  We want to be able to ascribe patterns of activation to different behavioral experiences, and have some confidence about the strength of association for one or more regions to an experience.  When we have this understanding, then we can bring in other phenotypic, genomic, "omic" variables to better understand their impact on the brain --> brain function --> behavior.  When you think about it, every behavior exists on a dimension, but sometimes people have manifestation of a behavior that is just a little too far in one direction so that it interferes with their daily life in a bad way.  This is what we call neurological disorder. We can't fix it with a band-aid, we barely understand most of it, but we get hints that there are many factors involved, ranging from what you had for dinner, to the genetic stuffs you got from your parents.  Helping these people is contingent upon establishing this basic understanding of the brain in normal and aberrant states. We don't have one massive neuroimaging study (some call this "mega analysis") but a bunch of tiny ones, and so essential to this goal is synthesizing this crap-ton of studies.  Trust me, there is lots of crap in there.  In my own terms, some behaviors make us happy, and others make us unhappy and sick, and since we have this idea that most of our human experience is coming out of the sensient meat between our ears, we probably need a map of that stuff to even know where to start.

## Image-Based Meta-Analysis
Now we are faced with the current challenge - an image-based (meaning using the whole statistical brain map) method that best addresses the questions of the researcher.  How about just taking an average? Or applying another level of random-effects models? There are a few challenges here that I will address:
- random-effects modeling requires that we know the variance of our units that went into the analysis. I understand that some software and going to be integrating buttons to export this information (on this level, it would be akin to variance between subjects), but my PhD needs to be in < 2 years, and I just don't have time for that.
- reverse inference! There is this thing called "forward inference," which means that if we do a study about anxiety, and then we see activation in the amygdala, we can say "yeah, anxiety --> amygdala." Unfortunately, this kind of information can be used inappropriately. Another researcher may do a study, and have a person look at cats, and see "cats --> amygdala," and then say "WOW, look what I've discovered!, cats --> anxiety!"  This is called reverse inference - or seeing activation in a region, and ascribing a cognitive process.  It's kind of irresponsible to do this because the reality is that a single brain region can be activated by many cognitive processes, so seeing the activation does NOT imply the cognitive process.

So, if we use standard meta-analysis approaches, they don't include reverse inference. But we really need to. If you read about the CBMA approaches, all this permutation-ing is really just producing an inference that says something is different from a null hypothesis. They can say something about common activation patterns for a set of studies, but they cannot produce reverse inferences.  In summary: we want to know what is the probability of this region --> cognitive process, or the likelihood of a result.  And this is important because individual studies are underpowered (small sample size).  We could arguably do the same thing with a mixed effects model, but what we don't have are the different variances that went into the model.

# Bayesian Inference
It is based on probabilities, which comes down to modeling uncertainty in our beliefs.  

 - p(A): is called the "prior" - what we believe to be the probability before we get any new information.  I think that the strategy I will want is to use "empirical" priors, meaning I learn them from the data.  In this case, more data is better.  But I also learned that priors aren't super important as long as the model is reasonable.  This might be the overall probability of some level of activation in a brain region.

> "When the facts change, I change my mind. What do you do, sir?"

 - p(A|X): is the updated belief, and we call this the "posterior probability."  This may be the probability of some level of activation in a brain region GIVEN some new data. The updated belief we can calculate with Bayes Formula:

![Bayes formula](http://upload.wikimedia.org/math/d/9/2/d92e290c66d423e4798a22a3690cbd31.png)

Vanessa interpretation: I start out stupid about everything, and update my beliefs as I get new evidence.  There is never a true answer, only something that we merge on based on lots of evidence. I think the idea is that some evidence may be better than other evidence (and maybe there are ways to reflect this with weighting it) but if you have enough data, you can still merge on pretty good answers.

Now I want to try using python to plot a sequence of posterior probabilities as we observe coin flips. This example is from [Cam Davidson's Notebook](http://nbviewer.ipython.org/github/CamDavidsonPilon/Probabilistic-Programming-and-Bayesian-Methods-for-Hackers/blob/master/Chapter1_Introduction/Chapter1.ipynb)

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
figsize(11, 9)

import scipy.stats as stats

dist = stats.beta

# Here we are make a range of trials
number_of_trials = [0, 1, 2, 3, 4, 5, 8, 15, 50, 500]

# Here we are generating bernoulli random variable with size equal t the largest number of trials
data = stats.bernoulli.rvs(0.5,size=number_of_trials[-1])

# Create a line from 0 to 1, break it into 100 pieces
x = np.linspace(0,1,100)

# For each of our trials (k is the count, N is the actual number)
for k, N in enumerate(number_of_trials):
  sx = plt.subplot(len(number_of_trials) / 2, 2, k + 1)
  plt.xlabel("$p$, probability of heads") \
    if k in [0, len(number_of_trials) - 1] else None
  plt.setp(sx.get_yticklabels(), visible=False)
  heads = data[:N].sum()
  y = dist.pdf(x, 1 + heads, 1 + N - heads)
  plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
  plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
  plt.vlines(0.5, 0, 4, color="k", linestyles="--", lw=1)
  leg = plt.legend()
  leg.get_frame().set_alpha(0.4)
  plt.autoscale(tight=True)

plt.suptitle("Bayesian updating of posterior probabilities",y=1.02,fontsize=14)
plt.tight_layout()
```

Lord, python makes beautiful plots. Let's talk about what these curves mean.  The posterior probabilities, which represent our belief in getting a head after observing some number of outcomes (generated from grabbing a value from a bernoulli distribution), are represented by the curves.  The width of the curves is like uncertainty. For example, when we start and have made no tosses, the probability of heads could be anything! (The plot is uniformly distributed between 0 and 1).  As we flip and add more evidence (data), we start to adjust this belief, and get more confident that the probability of a head is somewhere in the middle (50%).  Spoiler alert, given a fair coin, that kind of makes sense :)  It's notable that the plots aren't ALWAYS peaked at 0.5, for example look at 50 tosses! Change might have given us 15 heads and 35 tails. It happens! The idea is that if we approach larger numbers (infinity!), we might have other instances of getting, say 35 *heads* and 15 tails, and so we asymptote around the probability of 0.5.

Also remember that the values above were generated from a model, Mr. Bernoulli.  This is a "parametric" approach.  What I want to do is more empirical, and use data to update my beliefs. There will be no formal model, like Gaussian, Poisson, etc., so we can call this "non-parametric".

# Image-Based Meta-Analysis in Bayesian Framework
I am going to first define some things. 

 - A: as the event that a particular brain region is associated with a cognitive process.  
 - X: as the event of observing some level of activation in voxel b of the brain.
 
This means that the prior probability that a level of activation in a brain region is indicative of the cognitive process can be:

> P(A|X)
> P(brain region is associated with cognitive process|we see level of activation in voxel b)

There are a few details here that I want to touch on. I am being specific in saying "region" in part A, and "voxel" in X. A reflects the question, and X reflects the data.  Researchers (I've noticed) are super excited about the highest resolution of data, but they ask questions in the context of regions.  They may add some level of specificity (eg, "anterior insula"), but you rarely see someone say "voxel X Y Z." For this reason, I want my model to capture regional probabilities, but be based on voxel-wise activation. I don't know how I'm going to do this yet, but it's what I want.  And of course we will use standard regions (ontology) and allow for choice to choose a custom set of regions (brain atlases).

If we remember Bayes Formula, to answer this question we need to know:

> P(X|A)
> P(we see level of activation in voxel b|brain region is associated with cognitive process)

What is the probability of seeing that level of activation in our voxel b GIVEN that the brain region is associated with a cognitive process? Holy crap, I think we could figure this out from our data! Let's say that we just have a database with two kinds of study "group" statistical maps: one type of study has people seeing cats, and the other one has them seeing dogs. We now want to know the probability of seeing a Z score of range [1..2] in voxel [1,2,3] given that the brain region is associated with seeing cats? Well arguably we have a whole database of evidence (Z scores) for every single voxel, and every single brain region, being associated with seeing cats.  We can subset our database to the cat studies, do some tricks (I will discuss this part later) to calculate a probability of a level of activation for each region, and then plug it in.  Further, we could do the same thing for seeing dogs, and then compare if the brain region is just specific to cats, or both.

> P(X)
> P(we see level of activation in voxel b)

I don't (think) this is hard either.  This is doing the same kind of thing, but over *all* the data, both cats and dogs.  It's the overall probability of seeing activation in that voxel regardless of the cognitive task.  This is how reverse inference can come in - because if the overall probability of seeing activation in voxel b is super high, it's probably not a brain region specific to cats.  If the probability is low, but P(X|A) is high, we might have found the cat-epicenter. A good example might be with visual cortex.  Seeing a cat would activate visual cortex, as would anything. But, if there is a "cat-sensing" specific part of the brain, seeing the other stuff would not activate this part.  But we have a problem.

### Publication Bias
The brain maps that are published, or the ones that people will release, have a lot of bias.  Nobody is going to publish (or maybe even want to share) a map for a task with no interestingly high ("significant") values.  This means that we will likely be over-estimating our probabilities, and minimally need to keep this in mind as the method is developed. Back to piecing apart Bayes Formula in the context of IBMA.

### But we need to make a *formula*
If we are to have an algorithm that can calculate P(X) from the database, that is going to mean starting with one image of the database, and then taking our Z score values from a second image and using them to update our beliefs.  This means that what we really need is a formula for P(X) (the probability of seeing activation in voxel b) that can calculate this value

 - from things we already know (other variables, discussed above, from Bayes Formula
 - from new evidence!

To do this, let's break P(X) into two pieces:  

> P(X) = P(X and A) + P(X and not A)
> P(we see level of activation in voxel b) = P(we see level of activation in voxel b AND seeing cats) + (P we see level of activation in voxel b AND not seeing cats (aka, dogs))

and this gets reduced to:

> P(X) = P(X|A)p + P(X|~A)(1-p)

Where did this little p come from? Don't freak out, that is our prior probability P(A) but now we are representing it as variable p.  This formula is saying: "We can figure out the probability of seeing level of activation in voxel b by adding up:

- the probability of seeing activation in voxel b given region is associated with cats (we figured this out already) multipled by our prior probability.
- the probability of seeing activation in voxel b given the region is not associated with cats (so here we look at the dogs) and this gets multipled by (1-prior probability)

Remember that these funny looking values are going to mean summing up a bunch of probabilities, and are going to result in an actual number.  The way that we use each brain map and get the voxels to say something about a threshold for a region is not trivial (and I will need to think about this).  However for now, let's just make some numbers up. We will say after extensive analysis we determine:

- P(X|A)
- P(we see level of activation in voxel b|brain region is associated with seeing cats)

We calculate this to be 0.6, so we see activation at some level in broadly 60% of voxels b in our current cat sample.

- P(X|~A)

Same thing, but for cases when we aren't looking at cat data. Let's say that looking at cats only hits visual cortex, and that is also equally hit by seeing dogs, so the probabilities are pretty similar, we calculate as 0.55. Now we have a formula!

(0.6 * p) / ((0.6 * p) + 0.55(1-p))

reduces to:

0.6p / (1.15p + 0.55)

This is a function for the posterior probability, P(A|X), the probability that the region is associated with seeing cats given activation in the region, based on a previous belief! I'm so excited! Let's plot it!

```python
figsize(12.5, 4)
p = np.linspace(0, 1, 50)
plt.plot(p, 0.6 * p / (1.15*p + 0.55), color="#348ABD", lw=3)
# plt.fill_between(p, 2*p/(1+p), alpha=.5, facecolor=["#A60628"])
plt.scatter(0.1, 0.6 * (0.1) / 0.665, s=140, c="#348ABD")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel("Prior, $P(A) = p$")
plt.ylabel("Posterior, $P(A|X)$, with $P(A) = p$")
plt.title("Is the brain region associated with seeing cats?")
```

What this line shows us is our updated belief (y axis) based on our prior belief (x axis) of some brain region being associated with seeing cats.  What we see is that, even when our prior gets large, the calculation of the posterior sort of asymptotes around 0.4.  Why? Because we modeled the probability of seeing activation in the brain region for the rest of the experiences out there (eg, seeing dogs) as almost equally likely. The dot at 0.1 is just an example of, if set a prior of 0.1, the updated belief that the brain region is associated with cats is about the same. What about the other guy, the 1-p? This would be the probability of the brain region not being associated with seeing cats.

```python
figsize(12.5, 4)
colours = ["#348ABD", "#A60628"]

prior = [0.10, 0.90]
posterior = [1. / 3, 2. / 3]
plt.bar([0, .7], prior, alpha=0.70, width=0.25,
        color=colours[0], label="prior distribution",
        lw="3", edgecolor=colours[0])

plt.bar([0 + 0.25, .7 + 0.25], posterior, alpha=0.7,
        width=0.25, color=colours[1],
        label="posterior distribution",
        lw="3", edgecolor=colours[1])

plt.xticks([0.20, .95], ["Seeing Cats", "Not Seeing Cats"])
plt.title("Prior and Posterior probability of region associated with seeing cats")
plt.ylabel("Probability")
plt.legend(loc="upper left");
```
