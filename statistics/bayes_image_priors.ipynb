# Establishing priors for Image Based Meta Analysis (IBMA)

Previously, we reviewed [IBMA in the context of Bayesian inference](http://nbviewer.ipython.org/github/manbat/neuro-info/blob/master/statistics/bayes_for_ibma_intro.ipynb), and I determined the following:

 - the researcher asks questions at the level of brain regions
 - we want to work on the level of voxels
 - our priors will not come from distributions, they will come from data (empirical)

## We need to put our result in context of what came before it
And for a quick reminder of the "big picture" goal of this meta analysis:

 - increase statistical power for underpowered studies
 - find consistent activation of regions across studies
 - reverse inference: activity brain region --> cognitive processes

## We need Image Based Meta Analysis that includes reverse inference
We want to use predictive analyses to estimate (conditional on brain activation) the cognitive process or task that gave rise to our data. We want confident associations between cognitive processes and brain function so that we can better study and develop behavioral therapies. To do this, we need a meta-analysis for each class of task or cognitive-process "paradigm." We can't rely on coordinate based meta-analysis approaches because 1) they only infer effects that have consistent spatial overlap, 2) do not reflect the changing data landscape of having whole brain statistical maps.

## We need a Bayesian approach
We aren't close enough to also having people publicly share subject-level variances that would make another level of mixed-effects analysis possible.

```python
min = -4.5
max = 3.5
steps = ((abs(min)+max)*2)+1
breaks = np.linspace(min,max,num=steps,retstep=True)
print breaks # second value is the step size
breaks = breaks[0]
```

Ok, so I want to know the P(voxel b is in range X|region).  The conditioning is done by reducing the map to the region (already done), and so we just need to count the number of voxels in the range, and divide by the total number for the region.

```python
# First let's make column names to correspond with the intervals, regions will be in rows
column_names = []
for s in range(0,len(breaks)-1):
  start = breaks[s]
  stop = breaks[s+1]
  name = "[%s,%s]" %(start,stop)
  column_names.append(name)
column_names
```

```python
import pandas as pd
    
priors = pd.DataFrame(columns=column_names)
# values[0] is the values only for that region (so we are conditioning on values)
rownum = 0
for region,values in regions.iteritems():
  scores = []
  for s in range(0,len(breaks)-1):
    start = breaks[s]
    stop = breaks[s+1]
    indices = np.intersect1d(np.where(values[0] >= start)[0],np.where(values[0] < stop)[0])
    prob = float(len(indices))/len(values[0])
    scores.append(prob)
  # Now add the vector of scores to the matrix, row name should be region
  priors.ix[region] = scores 
```

```python
def plot_region(region,labels,color):
  ind = np.arange(len(region[1]))
  fig, ax = plt.subplots()
  fig.set_figwidth(12)
  fig.set_figheight(4)
  plot = ax.bar(ind,height=region[1], width=0.99, color=color)
  ax.set_ylabel('probability of activation')
  ax.set_xlabel('threshold range')
  ax.set_title("Probability Distribution for %s" %(region[0]))
  ax.set_xticklabels(labels)
```
