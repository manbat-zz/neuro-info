```python
import numpy
middles = [numpy.round(x/2) for x in atlas.get_shape()] 
mr = atlas.get_data()

# Generate each of the views
axial = numpy.rot90(mr[:,:,middles[0]],1)
sagittal = numpy.rot90(mr[middles[1],:,:],1)
coronal = numpy.rot90(mr[:,middles[2],:],1)

# Take a look
pl.imshow(axial, interpolation='nearest')
```

Don't worry, I didn't get those right on the first try! I want to make an important point about doing any kind of interpolation, either for visualization or resampling, when using atlas images.  If you do "continuous" (linear) interpolation, it will introduce values into the images that were not there before, essentially messing up your atlas.  Be careful :)  Next, we want to get the different edges of the images. Let's try a bunch of different edge detectors to see if any look good, for example: [Canny Edge Detection](http://en.wikipedia.org/wiki/Canny_edge_detector), and [sobel filters](http://en.wikipedia.org/wiki/Sobel_operator), which both take advantage of image gradients.

```python
from scipy import ndimage
from skimage.filter import canny
from scipy.ndimage.filters import sobel
import pylab as plt

# We will have three subplots
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3))

# Here is the original image
ax1.imshow(axial, cmap=plt.cm.jet)
ax1.axis('off')
ax1.set_title('original image', fontsize=20)

# Let's see what a sobel filter looks like
sobel_filter = sobel(axial, axis=-1, output=None, mode='nearest', cval=0.0)

# Here is the sobel
ax2.imshow(sobel_filter, cmap=plt.cm.jet)
ax2.axis('off')
ax2.set_title('sobel filter', fontsize=20)

# Here is a canny (I think this will be best)
canny_filter = canny(axial, 1, 0.0, 0.0)

ax3.imshow(canny_filter, cmap=plt.cm.jet)
ax3.axis('off')
ax3.set_title('canny filter', fontsize=20)

fig.subplots_adjust(wspace=0.02, hspace=0.02, top=0.9,
                    bottom=0.02, left=0.02, right=0.98)
```

I think it's likely that the [canny filter](http://scikit-image.org/docs/dev/api/skimage.filter.html?highlight=canny#skimage.filter.canny) is going to be best, but we need to fine tune some things.  First, we will definitely want to do edge detection for each region separately, to make life easier for the algorithm.  Let's first separate the regions:

```python
regions = [1,3,6]
fig, axis = plt.subplots(nrows=1, ncols=3, figsize=(8, 3))

for r in range(0,len(regions)):
  axial_filtered = numpy.zeros(numpy.shape(axial))
  axial_filtered[axial == regions[r]] = 1
  region = canny(axial_filtered,1,0,0)
  axis[r].imshow(region, cmap=plt.cm.jet)
  axis[r].axis('off')
  axis[r].set_title('region %s' %(regions[r]), fontsize=20)

fig.subplots_adjust(wspace=0.02, hspace=0.02, top=0.9,
                    bottom=0.02, left=0.02, right=0.98)
```

```python
import cairo
import rsvg
import math
from scipy.spatial.distance import pdist, squareform

# Just use one of the regions for now - the hardest one
axial_filtered = numpy.zeros(numpy.shape(axial))
axial_filtered[axial == 6] = 1
region = canny(axial_filtered,1,0,0)

# For now we will write svg to file
fo = file('/home/vanessa/Desktop/test.svg', 'wb')

# Set up the "context" - what cairo calls a canvas
width, height  = numpy.shape(region)
surface = cairo.SVGSurface (fo, width, height)
ctx = cairo.Context (surface)
ctx.set_source_rgb (0.3, 0.2, 0.5) # Solid color

# Create distance matrix for points
x,y = numpy.where(region==1)
points = [[x[i],y[i]] for i in range(0,len(x))]
disty = squareform(pdist(points, 'euclidean'))

# This keeps track of which we have already visited
visited = []
current = points[0]
row = 0
visited.append(row)

#for i in range(0,5):
while len(visited) != len(points):
  thisx = current[0]
  thisy = current[1]
  ctx.move_to(thisx, thisy)
  # Find closest point, only include columns we have not visited
  distances = disty[row,:]
  distance_lookup = dict()
  # We need to preserve indices but still eliminate visited 
  for j in range(0,len(distances)):
    if j not in visited: distance_lookup[j] = distances[j]
  # Get key minimum distance key
  row = min(distance_lookup, key=distance_lookup.get)
  next = points[row]
  nextx = next[0]
  nexty = next[1]
  ctx.line_to(nextx, nexty)
  # Set next point to be current
  visited.append(row)
  current = next

ctx.set_line_width (0.5)
ctx.stroke()

## Do the deed.
surface.finish()
fo.close()
```

```python
# Just use one of the regions for now - the hardest one
axial_filtered = numpy.zeros(numpy.shape(axial))
axial_filtered[axial == 6] = 1
region = img_as_float(find_boundaries(axial_filtered)) # We aren't using Canny anymore...

# For now we will write svg to file
fo = file('/home/vanessa/Desktop/test2.svg', 'wb')

# Set up the "context" - what cairo calls a canvas
width, height  = numpy.shape(region)
surface = cairo.SVGSurface (fo, width, height)
ctx = cairo.Context (surface)
ctx.set_source_rgb (0.3, 0.2, 0.5) # Solid color

# Segment!
segments_fz = felzenszwalb(region, scale=100, sigma=0.1, min_size=10)

# We need to skip 0, that is empty space
for r in range(1,len(numpy.unique(segments_fz))):
  # Yes, this is horrific, I'm ok with it for now :)
  cluster = numpy.zeros(numpy.shape(region))
  cluster[segments_fz==r] = 1
  # Create distance matrix for points
  x,y = numpy.where(cluster==1)
  points = [[x[i],y[i]] for i in range(0,len(x))]
  disty = squareform(pdist(points, 'euclidean'))

  # This keeps track of which we have already visited
  visited = []
  current = points[0]
  row = 0
  visited.append(row)
    
  # We need to remember the first point, for the last one
  fp = current

  while len(visited) != len(points):
    thisx = current[0]
    thisy = current[1]
    ctx.move_to(thisx, thisy)
    # Find closest point, only include columns we have not visited
    distances = disty[row,:]
    distance_lookup = dict()
    # We need to preserve indices but still eliminate visited 
    for j in range(0,len(distances)):
      if j not in visited: distance_lookup[j] = distances[j]
    # Get key minimum distance key
    row = min(distance_lookup, key=distance_lookup.get)
    next = points[row]
    nextx = next[0]
    nexty = next[1]
    # If the distance is more than N pixels, close the path
    # This smooths out some of the rough edges too
    if min(distance_lookup) > 10:
      ctx.line_to(fp[0],fp[1])
      ctx.set_line_width(1)
      ctx.close_path()
      fp = next
    else:    
      ctx.line_to(nextx, nexty)
      # Set next point to be current
    visited.append(row)
    current = next
    
  # Go back to the first point
  ctx.move_to(current[0],current[1])
  ctx.line_to(fp[0],fp[1])
  # Close the path
  ctx.set_line_width (1)
  ctx.stroke()
    
# Do the deed.
surface.finish()
fo.close()
```
